```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering


data = pd.read_csv("Mall_Customers.csv") 
data['Income'] = data[['Annual Income (k$)']]
data['Spending'] = data[['Spending Score (1-100)']] 
data = data[['Income', 'Spending']]

cure = AgglomerativeClustering(n_clusters=5, linkage='average', affinity='euclidean') 

labels = cure.fit_predict(data)

plt.scatter(data['Income'], data['Spending'], c=labels, cmap='rainbow', alpha=0.5, edgecolors='b') 
 
plt.xlabel('Income')
plt.ylabel('Spending')
plt.title('CURE Clustering of Mall Customers')
plt.show()
```
